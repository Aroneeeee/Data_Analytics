```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# 데이터 로드
data = pd.read_csv('data_1.csv')
```

```python
# 범주형 변수 'Attrition'을 0과 1로 인코딩 (0: No, 1: Yes)
label_encoder = LabelEncoder()
data['Attrition_encoded'] = label_encoder.fit_transform(data['Attrition'])
```

## 1. 연봉(MonthlyIncome)과 근속 연수(YearsAtCompany)를 기반으로 이직 여부에 대한 관계

* 분석에 사용할 주요 변수: 'MonthlyIncome', 'YearsAtCompany', 'Attrition_encoded'

```python
# 연봉과 근속 연수에 따른 이직률 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsAtCompany', y='MonthlyIncome', hue='Attrition', data=data, palette='coolwarm')
plt.title('Monthly Income and Years at Company by Attrition Status')
plt.xlabel('Years at Company')
plt.ylabel('Monthly Income')
plt.legend(title='Attrition', loc='upper left')
plt.show()
```

## 2. 직무 만족도와 근속 연수를 기반으로 이직 여부에 대한 분석

- 분석에 사용할 주요 변수: 'JobSatisfaction', 'YearsAtCompany', 'Attrition_encoded'

```python
# 직무 만족도와 근속 연수에 따른 이직률 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='JobSatisfaction', y='YearsAtCompany', hue='Attrition', data=data)
plt.title('Job Satisfaction and Years at Company by Attrition Status')
plt.xlabel('Job Satisfaction')
plt.ylabel('Years at Company')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

## 3. 연령대, 부서별, 초과근무에 따른 이직 여부 분석

- 분석에 사용할 주요 변수: Age, Department, Overtime, Attribution

### 1) 연령대에 따른 이직 여부 분석

```python
# 연령대별 이직 여부 분석을 위해 연령을 그룹화
bins = [18, 25, 35, 45, 55, 65]  # 연령 그룹 설정
labels = ['18-24', '25-34', '35-44', '45-54', '55-64']
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# 연령대별 이직률 시각화
plt.figure(figsize=(10, 6))
sns.countplot(x='AgeGroup', hue='Attrition', data=data, palette='Set2')
plt.title('Attrition by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

### 2) 부서별 이직 여부 분석

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=data, palette='Set1')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

### 3) 초과근무에 따른 이직 여부 분석

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='OverTime', hue='Attrition', data=data, palette='Set3')
plt.title('Attrition by OverTime')
plt.xlabel('OverTime')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```
